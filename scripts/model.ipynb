```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python

# this function is for the fully connected layer
def connected_layer(output_class):
    layer = tf.keras.Sequential([tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(output_class,activation='softmax')])
    
    return layer


# this functions resizes the training images to 224,224,3 before paasing it to the model for training   
def processing_layer(in_shape):
    process_layer = tf.keras.Sequential([#tf.keras.layers.Resizing(224,224,input_shape=in_shape),
                                          tf.keras.layers.Rescaling(1./255, input_shape = in_shape),
                                          tf.keras.layers.RandomFlip("horizontal_and_vertical"),
                                          tf.keras.layers.RandomRotation(0.2)
                                          ])
    return process_layer
    
```

```python
# this is the instance of the model, just call this class to instantiate this model and pass the neccessary methods to build a model 
# no need to hard code the architecture from scratch, just pass the training dataset and you have a new model
class Plant_Model:
    def __init__(self,in_shape=(224,224,3),output_class=40):
        self.in_shape = in_shape
        self.output_class = output_class
        self.trained_model = build_model()
        self.weights = save_weights()
        
    def build_model(self):
        self.model = tf.keras.Sequential([ 
            processing_layer(self.in_shape),
            tf.keras.layers.Conv2D(32,(3,3), activation='relu'),
            tf.keras.layers.MaxPooling2D(2,2),
            tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
            tf.keras.layers.MaxPooling2D(2,2),
            tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
            tf.keras.layers.MaxPooling2D(2,2),
            tf.keras.layers.Flatten(),
           # tf.keras.layers.dropout(0.5)
            connected_layer(self.output_class)

        ])
        self.model.compile(optimizer="adam",
                      loss="sparse_categorical_crossentropy",
                      metrics=['accuracy'])
        return self.model
    
    def save_weights(self,path):
        self.model.save(path)
        
```
